```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import math 
%matplotlib notebook
def GradosToRadians(grados):
    radianes = (math.pi * grados) / 180
    return radianes
def Geografic(geo):
    pen = (0-180)/(90-(-90))
    delta = (pen * geo)+90
    return delta
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
phi, theta = np.mgrid[0 : 2 * np.pi : 0.1, 0 : np.pi : 0.1] 
x = np.cos(phi) * np.sin(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(theta)

ax.plot_wireframe(x, y, z, color='green', rstride=2, cstride=2, linewidth=0.5, alpha=0.8)
Ciudades = [
    ['ParÃ­s', 48.8667, 2.3333],
    ['Lagos', 6.4433, 3.3915],
    ['BogotÃ¡', 4.5964, -74.0833],
    ['Roma', 41.8960, 12.4833],
    ['Melbourne', -37.8200, 144.9750],
    ['Nueva York', 40.6943, -73.9249],
    ['Los Angeles', 34.1139, -118.4068],
    ['MoscÃº', 55.7522, 37.6155],
    ['Londres', 51.5, -0.1167],
    ['Quito', -0.2150, -78.5001],
    ['Tokio', 35.6850, 139.7514],
    ['Nairobi', -1.2833, 36.8167],
    ['Santiago', -33.4500, -70.6670],
    ['Budapest', 47.5000, 19.0833],
    ['Vancouver', 49.2734, -123.1216],
    ['Atenas', 37.9833, 23.7333],
    ['BelÃ©n de ParÃ¡', -1.4500, -48.4800],
    ['Puebla', 19.0500, -98.2000]
    
]

for i in Ciudades:
    Latitud = i[1]
    Longitud = i[2]
    LatitudD =Geografic(Latitud)
    PX = np.cos(GradosToRadians(Longitud))*np.sin(GradosToRadians(LatitudD))
    PY = np.sin(GradosToRadians(Longitud)) * np.sin(GradosToRadians(LatitudD))
    PZ = np.cos(GradosToRadians(LatitudD))
    ax.scatter(PX,PY,PZ, color='blue', s=10)
    ax.text(PX,PY,PZ,i[0], color='black')
    
    
    
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Z')
ax.set_zlabel('Eje Y')

plt.show()
```
